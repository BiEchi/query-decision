```python
import pandas as pd

path = 'data/may_conversation_data_processed_with_prob.jsonl'
data = pd.read_json(path, lines=True)
```

```python
# set a threshold of 0.8, how many data points are above the threshold?
data[data['do_search_prob'] > 0.8].shape[0] / data.shape[0]

# plot the dist of data['do_search_prob']
data['do_search_prob'].plot.hist(bins=100)

# add a vertical line to show the threshold
import matplotlib.pyplot as plt
plt.axvline(x=0.8, color='red')

# add text to the plot, to show the percentage of data points above the threshold
plt.text(0.85, 50, '48.7%')
```

```python
data['do_search'] = data['do_search_prob'] > 0.82
```

```python
data['do_search'].value_counts()
```

```python
data.to_json('data/may_conversation_result.jsonl', orient='records', lines=True)
```
